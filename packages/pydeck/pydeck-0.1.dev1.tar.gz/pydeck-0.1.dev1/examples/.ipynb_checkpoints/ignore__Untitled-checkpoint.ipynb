```python
import pydeck as pdk
import requests
import pandas as pd
```

```python
A = pd.read_csv('https://data.sfgov.org/api/views/3mea-di5p/rows.csv?accessType=DOWNLOAD')
```

```python
A.head()
A['coordinates'] = A.apply(lambda x: [x['Longitude'], x['Latitude']], axis=1)
A.head()
```

```python
B = A[['coordinates', 'Address']]
# B = B[B['Address'] == '60 SPEAR ST']
B.head()
```

```python
C = pd.read_csv('https://data.sfgov.org/api/views/j2j3-acqj/rows.csv?accessType=DOWNLOAD')
C.head()
```

```python
D = C.merge(B, left_on='Building Address', right_on='Address')
# D = D[['Building Address', 'coordinates']]
COLS = [x for x in D.columns if 'kgCO2e' in x or 'Building Address' in x or 'coordinates' in x]
D['kgCO2e/ft2'] = D[COLS].apply(lambda x: x['2017 Total GHG Emissions Intensity (kgCO2e/ft2)'] or \
             x['2016 Total GHG Emissions Intensity (kgCO2e/ft2)'] or \
             x['2015 Total GHG Emissions Intensity (kgCO2e/ft2)'], axis=1)
E = D[['kgCO2e/ft2', 'coordinates', 'Building Address']]
E = E.dropna()
E['quantile'] = pd.qcut(E['kgCO2e/ft2'], 3, labels=["good", "medium", "bad"])
E.head()
```

```python
COLOR_LOOKUP = {
    'good': [0, 255, 0],
    'medium': [255, 255, 0],
    'bad': [255, 0, 0]
}
E['quantile_color'] = E.apply(lambda x: COLOR_LOOKUP[x['quantile']], axis=1)
E.head()
```

```python
df_dict = E.to_dict(orient='records') # Convert to a Python dictionary
```

```python
layer = pdk.Layer(
    'ScatterplotLayer',
    data=df_dict,
    get_position='coordinates',
    opacity=0.5,
    get_radius=40,
    get_color='quantile_color')
view_state = pdk.ViewState(
    zoom=11,
    latitude=37.74139927315054,
    longitude=-122.4529266357422
)
view = pdk.View('MapView', True)
b = pdk.Deck(layers=[layer], initial_view_state=view_state, views=[view])
b.map_style = None
b.show()
```
